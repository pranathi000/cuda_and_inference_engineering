```python
import torch
import torch.nn as nn
```

```python
model  = nn.Linear(512,512)
```

```python
# This layer has 512x512 = 262144 parameters
# In FP32: 262144 * 4 bytes = ~1 MB
```

```python
print(
    "Memory used:",
    model.weight.element_size() * model.weight.nelement(),
)
```

1048576÷10242=1 MB

```python
model_fp16 = model.half()
```

```python
print(
    "Memory used:",
    model_fp16.weight.element_size() * model_fp16.weight.nelement(),
)
```

```python
# Create input — must also be FP16 to match model
x = torch.randn(1, 512).half()
# batch size 1, 512 features, FP16
```

```python
with torch.no_grad():
    output = model_fp16(x)
```

```python
print("Output shape:", output.shape)
print("Output dtype:", output.dtype)
```

```python
model_bf16 = model.bfloat16()
x_bf16 = torch.randn(1, 512).bfloat16()

with torch.no_grad():
    output_bf16 = model_bf16(x_bf16)

print("\nBF16 output dtype:", output_bf16.dtype)
```
